```python
# Now we aim to predict some continuous numerical values

# Importing the libraries we need for the ML model we'll build
import numpy as np # for arrays 
import matplotlib.pyplot as plt # for charts
import pandas as pd # to pre-process our dataset & import them
```

```python
# Importing a dataset
dataset = pd.read_csv("Salary_Data.csv")
x = dataset.iloc[:, :-1].values # minus the last column (independent variables) (: ==> range)
y = dataset.iloc[:, -1].values # -1: the index of the last column (dependent variable) (we don't wont range)
```

```python
print(x)
```

```python
print(y)
```

```python
# Splitting the dataset into Training set and Test set
# Making 2 seperate tests
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 1) # random_state: to randomise train & test sets
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
# Training the "Simple Linear Regression Model" on the Training Set
# We'll use sklearn library to get access to a certain module(for our linear model)
# From this module we'll call a certain class called linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() # it does not need any parameters, directy creates a SLR model
regressor.fit(x_train, y_train) # fit: method for training our model
```

```python
# Predicting the Test Results
y_predicted = regressor.predict(x_test) # result into a vector
```

```python
# Visualising the Training Set results
plt.scatter(x_train, y_train, color = 'red') # that will allow us to put the red points(x,y) in the 2D
plt.plot(x_train,regressor.predict(x_train), color = 'blue') # our model we'll be a linear line
plt.title("Salary vs Experience (Training set)")
plt.xlabel('Years of experience') # name of the X-axis
plt.ylabel('Salary earned') # name of the Y-axis
plt.show() # Display the graphic
```

```python
# Visualising the Test set results
plt.scatter(x_test, y_test, color = 'green') # that will allow us to put the green points(x,y) in the 2D
plt.plot(x_train,regressor.predict(x_train), color = 'blue') # we'll get the same regression line(because we have a linear model)
plt.title("Salary vs Experience (Test set)")
plt.xlabel('Years of experience') # name of the X-axis
plt.ylabel('Salary earned') # name of the Y-axis
plt.show() # Display the graphic
```

```python
# Making a simple prediction
# Example: the salary of an employee with 12 years of experience

print(regressor.predict([[12]])) # double brackets: because the predict method expects a 2D array array as input
# 12==> scalar, [12]==> 1D, [[12]]==> 2D
```

```python
#  Getting thr final linear regression equation with the values of coefficients
print(regressor.coef_)
print(regressor.intercept_)
```
